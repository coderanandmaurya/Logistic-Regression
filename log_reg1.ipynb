```python
# import all lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,
                        n_redundant=0,n_classes=2,n_clusters_per_class=1,
                        random_state=42,class_sep=30)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],c=y,s=100)
plt.show()
```

```python
# perceptron
def perceptron(x,y):
  x=np.insert(x,0,1,axis=1)
  weight=np.ones(x.shape[1])
  lr=0.1

  for i in range(200):
    j=np.random.randint(0,100)
    y_hat =segmoid(np.dot(x[j],weight))
    weight=weight+lr*(y[j]-y_hat)*x[j]
  return weight[0],weight[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def segmoid(z):
  return 1/(1+np.exp(-z))
```

```python
intercept_,coef_=perceptron(x,y)
print(coef_,intercept_)
```

```python
m=-coef_[0]/coef_[1]
c=-intercept_/coef_[1]
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input +c
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_input,y_input,color='red')
plt.scatter(x[:,0],x[:,1],c=y,s=100,cmap='winter')
plt.ylim(-3,2) # used to get or set the y-limits of the current axes.
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
